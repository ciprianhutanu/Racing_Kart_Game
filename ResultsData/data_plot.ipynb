```python
import json
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
with open('lap_data.json', 'r') as file:
    data = json.load(file)

data = data['kartDataList']

times = [entry['time'] for entry in data]
positions = [entry['position'] for entry in data]
speeds = [entry['speed'] for entry in data]
drift = [entry['driftUsage'] for entry in data]

x = [pos['x'] for pos in positions]
y = [pos['y'] for pos in positions]
```

```python
ractrack = mpimg.imread('Helpers/racetrack.png')

plt.imshow(ractrack, extent=[0, 1, 0, 1], aspect='auto')
plt.plot(x, y, color='black', label='Racing Line')

plt.savefig('Results/race-line.png')
plt.show()
```

```python
plt.plot(times, speeds, color='red')
plt.xlabel('Timp (s)')
plt.ylabel('Vitez')

plt.fill_between(times, speeds, color='red', alpha=0.2)

plt.savefig('Results/lap-speed.png')
plt.show()
```

```python
plt.plot(times, drift, color='orange')
plt.xlabel('Timp (s)')
plt.ylabel('Utilizare Drift')

plt.fill_between(times, drift, color='orange', alpha=0.2)

plt.savefig('Results/drift-usage.png')
plt.show()
```

```python
plt.plot(times, speeds, color='red')
plt.plot(times, drift, color='orange')
plt.xlabel('Timp (s)')
plt.ylabel('Vitez / Utilizare Drift')

plt.fill_between(times, speeds, color='red', alpha=0.2)
plt.fill_between(times, drift, color='orange', alpha=0.2)

plt.savefig('Results/speed-drift-lap.png')
plt.show()
```

```python
def train_file_open(filename: str):
    with open(filename, 'r') as file:
        data = json.load(file)

    step = [entry[1] for entry in data]
    value = [entry[2] for entry in data]

    return step, value
```

```python
step, value = train_file_open('KartTraining1_Car Driver_cumulative_reward.json')

plt.plot(step, value, color='blue')
plt.xlabel('Pas')
plt.ylabel('Recompens')

plt.fill_between(step, value, color='blue', alpha=0.2)

plt.savefig('Results/cumulative-reward.png')
plt.show()
```

```python
step, value = train_file_open('KartTraining1_Car Driver_episode_length.json')

plt.plot(step, value, color='green')
plt.xlabel('Pas')
plt.ylabel('Lungime episod')

plt.fill_between(step, value, color='green', alpha=0.2)

plt.savefig('Results/episode-length.png')
plt.show()
```

```python
step, value = train_file_open('KartTraining1_Car Driver_policy_loss.json')

plt.plot(step, value, color='red')
plt.xlabel('Pas')
plt.ylabel('Policy loss')

plt.fill_between(step, value, color='red', alpha=0.2)

plt.savefig('Results/policy-loss.png')
plt.show()
```
